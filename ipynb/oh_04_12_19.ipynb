### Уменьшение площади под кривой спектра при уменьшении числа функциональных групп

```python
r_start = 780
r_end = 1470
a_goh3 = scipy.integrate.quad(res[0][2].func, r_start, r_end)[0]
a_104oh6 = scipy.integrate.quad(res[1][2].func, r_start, r_end)[0]

for struct_name, fg_name, func in res:
    print(struct_name, fg_name, ': площадь в диапазоне', r_start, '-', r_end, 'см-1:', 
          scipy.integrate.quad(func.func, r_start, r_end)[0])
    
print('Уменьшение площади на :', round((a_goh3 - a_104oh6), 2))
print('В % от исходной:', round((a_goh3 - a_104oh6) / a_goh3 * 100, 2), '%')
```

В структуре goh3 8 OH групп, а в структуре 104oh6 - 6. Таким образом, при уменьшении числа функциональных групп на (8 - 6) / 8 * 100% = 25% площадь под спектром OH групп уменьшается на примерно 26% в диапазоне частот 780 - 1470 см-1
