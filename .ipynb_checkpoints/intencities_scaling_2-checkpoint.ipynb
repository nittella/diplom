```python
structure_list = ['benzene', 'acetone', 'acetic_acid', 'isobutanol', 'benzoic_acid']
create_structures(structure_list)
structs = load_structures(structure_list)
struct_name='benzene'
```

```python
calc_maxes = find_function_max(structs[struct_name].pm.get_spectrum_function().calculate, structs[struct_name].freqs, bounds=(900, 4000))
spectrum = Spectrum(struct_name + '.csv')
exp_int = []
for freq_max in calc_maxes.peak_position:
    exp_int.append(spectrum.to_integrated_mol_absorptivity(freq_max))

    
calc_maxes['exp_int'] = exp_int
calc_maxes['sf'] = calc_maxes.intencity / calc_maxes.exp_int
```

```python
def gauss_function(x, mu=10, sigma=1, intencity_0=50):
    return intencity_0 * np.exp( - (x - mu) ** 2 / 2 / sigma)

mu = 9.5
x = []
y = []
delta = 2
for i in np.arange(mu - delta, mu + delta, 0.1):
    x.append(i)
    y.append(gauss_function(i))
x = np.array(x)
y = np.array(y)

gf_max = gauss_function(mu)
sigma = np.sqrt((x - mu) ** 2 / 2 / np.log(gf_max / y))
sigma = sigma[~np.isnan(sigma)]
sigma = sigma[sigma != float('inf')]
sigma = sigma[(sigma > 0.1) & (sigma < 10)]
sigma.mean()
```
