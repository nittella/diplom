```python
def approximate(X, function, basis_function1, basis_function2):
    i_exp = function
    i_1 = basis_function1
    i_2 = basis_function2
    
    a = sum(i_exp ** 2)
    b = 2 * sum(i_exp * i_1)
    c = 2 * sum(i_exp * i_2)
    d = sum(i_1 ** 2)
    e = 2 * sum(i_1 * i_2)
    g = sum(i_2 ** 2)
    
    determinant = 4 * d * g - e ** 2
    x_ex = (2 * b * g - c * e) / determinant
    y_ex = (2 * d * c - b * e) / determinant
    error = error_func(x_ex, y_ex, a, b, c, d, e, g)
    
    print(d, e, g, determinant)
    return error, x_ex, y_ex
```

```python
basis_function1 = GaussFunction(1, 0.2).calculate
basis_function2 = GaussFunction(2, 0.5).calculate
function = (GaussFunction(2, 0.2) + 2 * GaussFunction(3, 0.5)).calculate

x = np.arange(0, 10, 0.01)
plt.plot(x, basis_function1(x), label='basis function 1')
plt.plot(x, basis_function2(x), label='basis function 2')
plt.plot(x, function(x), label='function for approximation')

plt.legend()
```

```python
error, w1, w2 = approximate(function, basis_function1, basis_function2, bounds=(0, 10), step=0.1)
print(error, w1, w2)
```

```python
approximated_function = (w1 * GaussFunction(1, 0.2) + w2 * GaussFunction(2, 0.5)).calculate

plt.plot(x, function(x), label='function for approximation')
plt.plot(x, approximated_function(x), label='approximated function')
```
