```python
def generate_features_set(atoms=['O', 'C', 'H']):
    all_features = set()
    for atom1 in atoms:
        for atom2 in atoms:
            atom_list_1 = [atom1, atom2]
            all_features.add(''.join(sorted(atom_list_1)))
            for atom3 in atoms:
                atom_list_2 = [atom1, atom2, atom3]
                all_features.add(''.join(sorted(atom_list_2)))
                for atom4 in atoms:
                    atom_list_3 = [atom1, atom2, atom3, atom4]
                    all_features.add(''.join(sorted(atom_list_3)))
    return all_features
print(len(generate_features_set()))
print(sorted([x for x in generate_features_set()], key=lambda x: len(x)))
```

```python
def analyze_mode_type_2(coordinates_1):
    all_features = generate_features_set()
    vector_1_ = dict()
    for feature in all_features:
        vector_1_[feature] = 0
    for value, complex_coordinate in coordinates_1:
        for coordinate in complex_coordinate.coordinates:
            coordinate.name.replace('Ru', 'H')
            atoms = ''.join(sorted([x for x in coordinate.name.replace('Ru', 'H')]))
            vector_1_[atoms] += abs(value) / len(complex_coordinate.coordinates)
    # print([x for x in vector_1.values()])
    # for key in ['TORS', 'BEND', 'OUT', 'STRE', 'TORS_O', 'BEND_O', 'OUT_O', 'STRE_O']:
    #     vector_1_.pop(key)
    return [x for x in vector_1_.keys()], [x for x in vector_1_.values()]
```

Попробуем сгруппировать частоты по косинусным расстояниям.
Найдем расстояние от каждого вектора до всех остальных и отсортируем их по возрастанию

```python
def clusterize_freq(structure_name, step, data=None, normalize=True):
#     тут находим все расстояния между векторами и сортируем их по возрастанию
    freqs = ped_dict[structure_name].get_freqs()
    data_matrix = []
    labels = []
    for freq in freqs:
        labels, vector = analyze_mode_type_2(ped_dict[structure_name].get_coord_by_freq(freq))
        data_matrix.append(vector)
        
# стандартизуем значения https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html
    if normalize:
        data_matrix_normalized = []
        for column in np.array(data_matrix).T:
            column = column.reshape(-1, 1)
            scaler = StandardScaler()
            scaler.fit(column)
            column = scaler.transform(column)
            column = np.squeeze(column, axis=1)
            data_matrix_normalized.append(column)
        data_matrix = np.array(data_matrix_normalized).T
#     print(data_matrix_normalized)
#     print(data_matrix)
    
    distances_ = dict()
    for i, vector_1 in enumerate(data_matrix):
        distances_[freqs[i]] = []
        for j, vector_2 in enumerate(data_matrix):
            if i != j:
                distances_[freqs[i]].append((cosine(vector_1, vector_2), freqs[j]))
        distances_[freqs[i]].sort(key=lambda x: x[0])

#   а тут кластеризуем их по шагу step
    freq_for_groups = set(freqs)
    clusters = []
    clusters_matrix = []
    while len(freq_for_groups) != 0:
        new_cluster = []
        new_cluster_matrix = []
        stack = set()
        freq = freq_for_groups.pop()
        stack.add(freq)
        new_cluster.append(freq)
        new_cluster_matrix.append(data_matrix[freqs.index(freq)])
        while len(stack) != 0:
            for distance, freq_i in distances_[stack.pop()]:
                if distance < step:
                    if freq_i in freq_for_groups:
                        freq_for_groups.remove(freq_i)
                        if freq_i not in new_cluster:
                            new_cluster.append(freq_i)
                            new_cluster_matrix.append(data_matrix[freqs.index(freq_i)])
                        stack.add(freq_i)
        clusters.append(new_cluster)
        clusters_matrix.append(new_cluster_matrix)
        
# можно еще посмотреть дисперсию и среднее для каждого кластера по каждой координате
# будем делать это только если предан параметр data='all'
    if data == 'all':
#         print(*zip(labels, np.array(clusters_matrix[3]).T), sep='\n'
        return clusters, labels, clusters_matrix
    return clusters
```

Опишем для каждой структуры какие номера атомов кислорода к каким функциональным группам относятся

```python
oxygen_in_structure = {
    '104': {
        72: 'e',
        73: 'e',
        74: 'e'
    },
    '104oh6': {
        72: 'e',
        73: 'e',
        74: 'e',
        76: 'h',
        77: 'h',
        78: 'h',
        79: 'h',
        80: 'h',
        81: 'h'
    },
    'g104w3': {
        72: 'e',
        73: 'e',
        74: 'e',
        76: 'w',
        79: 'w',
        80: 'w'
    },
    '104oh6w6': {
        72: 'e',
        73: 'e',
        74: 'e',
        76: 'h',
        77: 'h',
        78: 'h',
        79: 'h',
        80: 'h',
        81: 'h',
        88: 'w',
        91: 'w',
        94: 'w',
        97: 'w',
        100: 'w',
        103: 'w',
    }
}
func_groups = ('epoxide', 'hydroxide', 'water', 'carboxide', 'keton')
```

```python
def analyze_mode_type_3(coordinates_1, structure_name):
    all_features = generate_features_set(atoms=['C', 'H', 'Oe', 'Oh', 'Ow'])
    vector_1_ = dict()
    for feature in all_features:
        vector_1_[feature] = 0
    for value, complex_coordinate in coordinates_1:
        for coordinate in complex_coordinate.coordinates:
            name_list = [x for x in coordinate.name.replace('Ru', 'H')]
            while 'O' in name_list:
                ind = name_list.index('O')
                o_number = coordinate.atoms[ind]
                name_list[ind] += oxygen_in_structure[structure_name][o_number]            
            atoms = ''.join(sorted(name_list))
            vector_1_[atoms] += abs(value) / len(complex_coordinate.coordinates)
    return [x for x in vector_1_.keys()], [x for x in vector_1_.values()]
    
```

## Подготовка матрицы

```python
def prepare_for_clustering(structure_name, normalize=True):
# делаем матрицу векторов
    freqs = ped_dict[structure_name].get_freqs()
    data_matrix = []
    labels = []
    for freq in freqs:
        labels, vector = analyze_mode_type_3(ped_dict[structure_name].get_coord_by_freq(freq), structure_name)
        data_matrix.append(vector)
        
# переносим в пандас        
    matrix_frame = pd.DataFrame(data_matrix)
    matrix_frame.columns = labels
    matrix_frame['freqs'] = freqs

# удаляем пустые столбцы
    matrix_frame = matrix_frame.loc[:, (matrix_frame != 0).any(axis=0)]
    new_lables = [x for x in matrix_frame.columns]
        
# стандартизуем значения https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html

    if normalize:
        scaler = StandardScaler()
        scaler.fit(matrix_frame)
        scaled_frame = scaler.transform(matrix_frame)
#         data_matrix_normalized = []
#         for column in np.array(data_matrix).T:
#             column = column.reshape(-1, 1)
#             scaler = StandardScaler()
#             scaler.fit(column)
#             column = scaler.transform(column)
#             column = np.squeeze(column, axis=1)
#             data_matrix_normalized.append(column)
#         data_matrix = np.array(data_matrix_normalized).T
    return scaled_frame, new_lables, freqs, scaler
```

Применим функцию ко всем структурам и будем хранить в словаре

```python
prepaired_data = dict()
for name in ['104', '104oh6', 'g104w3', '104oh6w6']:
    data = prepare_for_clustering(name)
    prepaired_data[name] = dict()
    prepaired_data[name]['matrix'] = data[0]
    prepaired_data[name]['labels'] = data[1]
    prepaired_data[name]['freqs'] = data[2]
    prepaired_data[name]['scaler'] = data[3]
```

Посмотрим по частотам

```python
from scipy.stats import variation

def build_clasters(labels, name):
    cluster_number = max(labels) + 1
    freq_clusters = [[] for x in range(cluster_number)]
    for i, n in enumerate(labels):
        freq_clusters[n].append(prepaired_data[name]['freqs'][i])
    rescaled_matrix = prepaired_data[name]['matrix']    
#     rescaled_matrix = prepaired_data[name]['scaler'].inverse_transform(prepaired_data[name]['matrix'])
    frame = pd.DataFrame(rescaled_matrix)
    frame.columns = prepaired_data[name]['labels']
    frame['cluster'] = labels
    
    clusters = []
    for n in range(cluster_number):
        clusters.append(frame.loc[frame['cluster'] == n])
    
    means_stds = []
    stds = []
    for cluster in clusters:
        means_stds.append((cluster.mean(axis = 0), cluster.std(axis = 0) / cluster.mean(axis = 0)))
        stds.append(cluster.std(axis = 0) / cluster.mean(axis = 0))
    
    return freq_clusters, clusters, means_stds, stds
```

```python
from math import isnan
def analize_cluster_decomposition(labels, name):
    freq_clusters, clusters, means_stds, stds = build_clasters(labels, name)
    total_std = 0
    cluster_number = max(labels) + 1
    for i in range(cluster_number):
        print('cluster', i)
        stds_sum = sum([x for x in stds[i] if not isnan(x)])
        total_std += stds_sum / len(freq_clusters[i])
        mean_stds_round = [x for x in zip(*map(lambda x: round(x, 2), means_stds[0]))]
        means_stds_frame = pd.DataFrame(means_stds[i][0].values)
        means_stds_frame.columns = ['means']
        means_stds_frame['stds'] = means_stds[i][1].values
        means_stds_frame['labels'] = means_stds[i][0].index
        print('sum std', stds_sum)
        print('std per element', stds_sum / len(freq_clusters[i]))
        print(*(clusters[i]['freqs']))    

#         print(means_stds_frame)
        print('best modes:')
        print(means_stds_frame[:-2].loc[means_stds_frame['means'][:-2] > 10])
        print()    
#         print(type(means_stds[i][0].values))
    print('TOTAL STD per element', total_std)
```

Сделаем оценку кластеризации:

The best value is 1 and the worst value is -1. Values near 0 indicate overlapping clusters. Negative values generally indicate that a sample has been assigned to the wrong cluster, as a different cluster is more similar.

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score
silhouette_score(prepaired_data['104']['matrix'], k_means.labels_)
# calinski_harabasz_score(matrix_1, labels_1) # пока непонятная метрика
```

Посмотрим на зависимость оценки от числа кластеров:

```python
# %matplotlib qt
def make_clusters_score_plot(name):
#     %matplotlib inline
    matrix, lables, freq, scaler = prepare_for_clustering(name)
    ns = [x for x in range(2, 70)]
    scores = []
    for n in ns:
        k_means = KMeans(n_clusters=n, random_state=0)
        k_means.fit(matrix)
        scores.append(silhouette_score(matrix, k_means.labels_))
    plt.plot(ns, scores, label=name)
    plt.xlabel('n_clusters')
    plt.ylabel('score')
    plt.legend()
```
