## Предварительное изучение

Загружаем данные структуры:

```python
create_structures(['gketoohw12'])
structs = load_structures(['gketoohw12'])
```

В структуре находтся следующие структурные элементы:
* 12 молекул воды;
* 14 OH групп (6 в центре и 8 на переферии);
* 2 кето-группы на переферии.

Посмотрим на расчетный спектр:

```python
fig, ax1 = plt.subplots(figsize=(20, 10))
structs['gketoohw12'].pm.spectrum_curve(axis=ax1)
structs['gketoohw12'].pm.spectrum_bar(axis=ax1)
structs['gketoohw12'].pm.fg_bar('water', axis=ax1)
plt.xlim(900, 4000)
plt.legend()
```

## Диапазон выше 3000 см-1
Посмотрим на расчетный спектр воды в этотм диапазоне. Специально взяли полуширину в два раза меньше, чтобы подробнее рассмотреть диапазон  3400 - 3650 см-1.

```python
fig, ax2 = plt.subplots(figsize=(20, 10))
structs['gketoohw12'].pm.fg_bar('water', axis=ax2)
structs['gketoohw12'].pm.fg_sum_curve('water', axis=ax2, sigma=15, annotate=True)[0]
plt.xlim(left=3000, right=3800)
plt.legend()
```

В таблице ниже приведены расчетные скорректированные моды, две внутренние координыты, внесшие значительный вклад в потенциальную энергию и гаусов пик, в который наибольший вклад дают эти моды. 

```python
water_curve = structs['gketoohw12'].pm.get_spectrum_function('water', sigma=15).calculate
water_points = structs['gketoohw12'].get_fg_freqs('water', corrected=True)
water_peaks_table = find_function_max(water_curve, water_points, bounds=(3000, 4000))
report_table = structs['gketoohw12'].get_report('water', corrected=True, freq_range=(3000, 4000))
report_table['gauss_max'] = water_peaks_table.peak_position
report_table['old_freq'] = structs['gketoohw12'].get_fg_freqs('water', corrected=False)[(structs['gketoohw12'].get_fg_freqs('water', corrected=True) < 4000) & (structs['gketoohw12'].get_fg_freqs('water', corrected=True) > 3000)]
report_table
```

Рассмтрим первый пик со значением 3714 см-1.

```python
report_table.drop(['gauss_max'], axis=1).loc[report_table.gauss_max == 3714.0]
```

Как видно из таблицы приведенной выше, таких частот всего 7. Основной вклад в потенциальную энергию вносит изменение вдоль OH связей различных молекул воды.

<img src="images/gketoohw12_H1.jpg" style="display: inline-block; max-height: 100%;">

Как видно из ррисунка выше (лучше псмотреть саму структуру в ChemCraft), протоны, связь которых с атомом кислорода вносит основной вклад в потенциальную энергию этих мод, не имеют никаких водородных связей.

Итоги:
* 3714 см-1 - нет водородных связей
* 3596, 3520, 3441 см-1 - слабая водородная связь (1.87 - 2.04 A)
* 3349 см-1 - в основном колебание гидроксильной группы, длина водородной связи протона молекулы воды 1.83 A
* 3249 см-1 - колебания ансамбля молекул воды и гидроксильных групп, длина водородной связи 1.74 A
* 3091 см-1 - колебания ансамбля молекул воды и гидроксильных групп, очень сильная водородная связь 1.67 А. Замкнутая цепочка валентная связь OH - водородная связь ОН колеблется взаимосвязано, основной вклад вносят колебания звеньев в начале и в конце цепочки (Колыбе́ль Ньютона).


## Диапазон  900 - 1700 см-1

```python
fig, ax2 = plt.subplots(figsize=(20, 10))
structs['gketoohw12'].pm.fg_bar('water', axis=ax2)
structs['gketoohw12'].pm.fg_sum_curve('water', axis=ax2, sigma=30, annotate=True)[0]
plt.xlim(left=900, right=1800)
plt.legend()
```

```python
water_curve = structs['gketoohw12'].pm.get_spectrum_function('water', sigma=30).calculate
water_points = structs['gketoohw12'].get_fg_freqs('water', corrected=True)
water_peaks_table = find_function_max(water_curve, water_points, bounds=(900, 1800))
report_table = structs['gketoohw12'].get_report('water', corrected=True, freq_range=(900, 1800))
report_table['gauss_max'] = water_peaks_table.peak_position
report_table['old_freq'] = structs['gketoohw12'].get_fg_freqs('water', corrected=False)[(structs['gketoohw12'].get_fg_freqs('water', corrected=True) < 1800) & (structs['gketoohw12'].get_fg_freqs('water', corrected=True) > 900)]
report_table
```

Итоги:
* 1653 см-1 - изменение углов HOH в молекулах воды.
* 1314 см-1 - изменение угла C-O-H, затрагивающее изменение угла между соседней молекулой воды и этой гидроксильной группой. Это гидроксильная группа, у которой протон сильно связан водородной связью с молекулой воды (1.45 А)
* 1032 см-1 - тау-колебания молекул воды, затрагивающие валентные колебания CO связи и колебания каркаса
